```python
## Define vaiables

i = 0.0002
L = 1020
weir_height = 2.3
weir_width = 100

H_e = 7.95
U_e = q_e / H_e
Uw_e = q_e / (H_e - weir_height)
```

```python
## Extreme Flow Conditions

Q_e = 660
q_e = Q_e / weir_width

Hs_e = 7.798
U_e = q_e / Hs_e
Uw_e = q_e / (Hs_e - weir_height)
Uloc_e = 2 * U_e

print("The flow velocity in extreme conditions is:", U_e, "m/s")
print("The local flow velocity in extreme conditions is:", Uloc_e, "m/s")
print("The flow velocity over the weir in extreme conditions is:", Uw_e, "m/s")

Fr_e = Uw_e / (9.81 * (Hs_e - weir_height))**0.5
print("The Froude number in extreme conditions is:", Fr_e)

r = 0.2
u_max_e = U_e * (1 + 3 * r)

print("The maximum flow velocity in extreme conditions is:", u_max_e, "m/s")
```

```python
## Find D50 downstream

from math import log10

shields = 0.03
chezy = 50
delta = 1.65

## Create loop to find D50

D50_new = 0.01 #initial guess
D50 = 0.02 #initial value to enter loop

while abs(D50_new - D50) > 0.00001:
    D50 = D50_new
    k = 2*D50
    C = 18*log10(12*Hs_e/k)
    D50_new = (1*U_e**2)/(1*shields*delta*C**2)

print("The median grain size before calibration is:", D50_new, "m")
print("The Chezy roughness is:", C, "m^(1/2)/s")

Ks = 1
Kv = 1.3

D50 = D50_new * (Kv**2 / Ks)

print("The calibrated median grain size is:", D50, "m")
```

```python
## Critical Flow

dh = i * L
hw_c = 2*dh # for critical flow
uw_c = (9.81*hw_c)**0.5
r = 0.2

Q=hw_c*uw_c*weir_width
print("The critical discharge over the weir is:", Q, "m^3/s")
h_downstream=hw_c + weir_height
print("The flow depth downstream of the weir is:", h_downstream, "m")
u_downstream=Q/(h_downstream*weir_width)
print("The flow velocity downstream of the weir is:", u_downstream, "m/s")
print("The flow depth over the weir is:", hw_c, "m")
print("The flow velocity over the weir is:", uw_c, "m/s")
print("The maximum flow velocity downstream is:", u_max_c, "m/s")
```

```python
## Find D50 on the weir
import numpy as np

D50 = 0.02 #initial guess

while abs(D50_new - D50) > 0.00001:
    D50 = D50_new
    k = 2*D50
    C = 18*log10(12*(hw_c)/k)
    D50_new = (1*uw_c**2)/(1*shields*delta*C**2)

print("The median grain size on the weir before calibration is:", D50_new, "m")

phi = 41
alpha = np.degrees(np.arctan(1/3))

Ks_c = np.sin(np.radians(phi - alpha))/np.sin(np.radians(alpha))
Kv_c = 1.3

D50_c = D50_new * (Kv_c**2 / Ks_c)

print("The calibrated median grain size on the weir is:", D50_c, "m")

hw_c/D50_c
```

```python
## Izbash

D_izbash = (Kv_c**2 * uw_c**2) / (Ks_c * 9.81 * delta * 2.9)

print("The Izbash diameter is:", D_izbash, "m")

hw_c/D_izbash
```
